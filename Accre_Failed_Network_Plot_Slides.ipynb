```python
c = ['red', 'yellow', 'thistle', 'thistle', 'gainsboro', 'skyblue', 'gold', 'gold','oldlace']

ACCOUNT={'EXITCODE':['1:0', '2:0', '127:0', '7:0', '13:0', '16:0', '126:0', '33:0', '9:0'],
          'FAILED': [321, 16,6,6,5,3,2,2,1]}
                 
Users = pd.DataFrame(ACCOUNT, columns=['EXITCODE', 'FAILED'])

plt.figure(figsize=(15,10))

plots = sns.barplot(x='EXITCODE', y='FAILED', data= Users, palette = c, edgecolor = 'black', alpha = .5)

for bar in plots.patches:

    plots.annotate(format(bar.get_height(), '.2f'),
                   (bar.get_x() + bar.get_width()/2,
                   bar.get_height()), ha = 'center', va = 'center',
                   size = 13, xytext = (0,8),
                   textcoords = 'offset points',
                  color = 'black')

plt.xlabel('Exitcode', size = 15)

plt.ylabel('Occurences', size = 15)

plt.title('Exitcode occurences', fontsize=15)

```
